```python
import pandas as pd

gestor_jornada = pd.read_excel('planilha_gestor_jornada.xlsx')
gestor_jornada.drop(columns=['Placa', 'NomeMacro', 'TipoMacro', 'PontoReferencia', 'KM'], inplace=True)
gestor_jornada.dropna(inplace=True)
gestor_jornada.sample(10)
```

```python
colaboradores = gestor_jornada["Nome"].unique()
# gestor_jornada["DataInicio"] = gestor_jornada["DataInicio"][:9]
# print(gestor_jornada["DataInicio"][0:9])
# gestor_jornada["DataFim"] = gestor_jornada["DataFim"][:9]
# print(gestor_jornada["DataFim"].value_counts())

new_colaboradores_df = pd.DataFrame(columns=["nome", "periodo", "total_horas", "escala", "disponibilidade"])

periodos = []
periodo = {
    "nome": "",
    "data_inicial" : None,
    "data_final": None,
    "total_horas": 0,
    "escala": 0,
    "disponibilidade": 0
}
for colaborador in colaboradores:
    colaborador_df = gestor_jornada[gestor_jornada["Nome"] == colaborador].copy()
    loop_index = []

    for index in range(len(colaborador_df)):
        loop_index.append(index)
        plus_index = index + 1 if index < len(colaborador_df) - 1 else index
        diff = colaborador_df.iloc[plus_index]["DataInicio"] - colaborador_df.iloc[index]["DataInicio"]
        if diff.days <= 1:
            periodo["data_inicial"] = colaborador_df.iloc[min(loop_index)]["DataInicio"]
            periodo["data_final"] = colaborador_df.iloc[index]["DataInicio"]
        else:
            periodos.append(periodo)
            loop_index = []
    
for item in periodos:
    print(item)
    # new_colaboradores_df.loc[len(periodos)] = item

# new_colaboradores_df
```
